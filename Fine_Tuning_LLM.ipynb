# Xử lý dữ liệu

```python
fpt['ticker'] = "FPT"
cmg['ticker'] = "CMG"
price_all = pd.concat([fpt, cmg], ignore_index=True)
price_all.rename(columns={"Date": "date"}, inplace=True)
price_all = price_all[['date','Closing Price','ticker']]
price_all
```

```python
price_all.info()
```

```python
price_all['date'] = pd.to_datetime(price_all['date'], format='%d/%m/%Y')
price_all['Closing Price'] = price_all['Closing Price'].str.replace(',','', regex=True).astype(int)
price_all.info()
```

```python
news_grouped = news.groupby(["date"])["title"].apply(lambda x: "; ".join(x)).reset_index()
news_grouped['date'] = pd.to_datetime(news_grouped['date'], format='%d/%m/%Y')
news_grouped
```

```python
df = price_all.merge(news_grouped, how='left', on=['date'])
df = df.sort_values(by=["ticker", "date"]).reset_index(drop=True)
df
```

```python
df.info()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

# Tải tokenizer và model
model_id = "wonrax/phobert-base-vietnamese-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=False)
model = AutoModelForSequenceClassification.from_pretrained(model_id)

def analyze_sentiment(text):
    try:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
        with torch.no_grad():
            outputs = model(**inputs)
            probs = F.softmax(outputs.logits, dim=-1).squeeze()
            score = float(probs.max())
            label_id = int(torch.argmax(probs))

            # Chuyển label id thành chuỗi tương ứng
            label_map = {0: "NEG", 1: "POS", 2: "NEU"}
            label_str = label_map.get(label_id, "Unknown")
            return label_str, score
    except Exception as e:
        print(f"Lỗi khi phân tích cảm xúc: {e}")
        return "Unknown", 0.0

df["sentiment_label"], df["sentiment_score"] = zip(*df["title"].map(analyze_sentiment))
df
```

```python
from datasets import Dataset
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer, Trainer, TrainingArguments
from transformers import DataCollatorForSeq2Seq
import torch
import numpy as np
from sklearn.metrics import mean_squared_error

# Tiền xử lý dữ liệu (thêm kiểm tra chuỗi rỗng)
inputs = []
targets = []

# Với mỗi ngày hôm nay, lấy giá của ngày thứ ba (3 ngày sau)
for i in range(len(df) - 3):  # Cần ít nhất 3 ngày sau
    today_row = df.iloc[i]
    today_price = today_row["Closing Price"]
    day1 = df.iloc[i + 1]["Closing Price"]
    day2 = df.iloc[i + 2]["Closing Price"]
    day3 = df.iloc[i + 3]["Closing Price"]

    # Tính delta (chênh lệch giá)
    delta1 = round(day1 - today_price)
    delta2 = round(day2 - today_price)
    delta3 = round(day3 - today_price)

    # Lấy cảm xúc từ dữ liệu
    sentiment_label = today_row["sentiment_label"]
    sentiment_score = today_row["sentiment_score"]

    # (Tuỳ chọn) Ánh xạ sang tiếng Việt cho dễ hiểu hơn
    label_map_vn = {"POS": "Tích cực", "NEG": "Tiêu cực", "NEU": "Trung lập", 'Unknown': 'Không có thông tin'}
    sentiment_text = label_map_vn.get(sentiment_label, "Không rõ")

    # Tạo input text bao gồm cảm xúc thật
    input_text = (
        f"Dự báo: Giá hôm nay của {today_row['ticker']} là {today_price}. "
        f"Tin tức: {today_row['title']}. Cảm xúc: {sentiment_text} (Điểm: {float(sentiment_score):.2f})"
    )

    # Target là delta
    target_text = f"{delta1}, {delta2}, {delta3}"

    # Bỏ dòng nếu thiếu thông tin
    if pd.notna(input_text) and pd.notna(target_text) and input_text.strip() and target_text.strip():
        inputs.append(input_text)
        targets.append(target_text)

# Tiền xử lý để thay thế nan bằng chuỗi rỗng
inputs = [input if input != 'nan' else '' for input in inputs]
targets = [target if target != 'nan' else '' for target in targets]

# Kiểm tra xem có chuỗi trống nào không
inputs = [input for input in inputs if input.strip() != '']
targets = [target for target in targets if target.strip() != '']

# Tạo dataset
dataset = Dataset.from_dict({"input": inputs, "target": targets})
dataset = dataset.train_test_split(test_size=0.1)

# Tải tokenizer và model gốc
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

def preprocess(batch):
    model_inputs = tokenizer(batch["input"], max_length=256, truncation=True, padding="max_length")

    labels = tokenizer(
        batch["target"],
        max_length=20,  # <-- cập nhật ở đây
        truncation=True,
        padding="max_length",
        text_target=batch["target"]
    )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_dataset = dataset.map(preprocess, batched=True)

# Đóng gói dữ liệu
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)

# Định nghĩa tham số huấn luyện
training_args = TrainingArguments(
    output_dir="./results",
    save_strategy="epoch",
    save_total_limit=1,
    logging_strategy="epoch",
    logging_steps=50,
    num_train_epochs=5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    learning_rate=3e-5,
    weight_decay=0.01,
    logging_dir="./logs",
    push_to_hub=False,
    report_to="none",
    disable_tqdm=True,
    metric_for_best_model="loss",
)

def compute_metrics(eval_pred):
    predictions, labels = eval_pred

    # Giải mã các chuỗi
    decoded_preds = tokenizer.batch_decode(predictions, skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    try:
        # Chuyển đổi về float (mỗi chuỗi là "x, y, z")
        pred_nums = [list(map(float, pred.split(','))) for pred in decoded_preds]
        label_nums = [list(map(float, label.split(','))) for label in decoded_labels]

        # Tính RMSE trung bình trên từng chuỗi
        rmses = []
        for pred, label in zip(pred_nums, label_nums):
            if len(pred) == len(label):
                rmse = mean_squared_error(label, pred, squared=False)
                rmses.append(rmse)

        return {"rmse": float(np.mean(rmses)) if rmses else None}

    except Exception as e:
        print("Lỗi trong compute_metrics:", e)
        return {}

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# Huấn luyện mô hình
trainer.train()

# Lưu mô hình và tokenizer sau huấn luyện
trainer.save_model("./trained_model")
tokenizer.save_pretrained("./trained_tokenizer")
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Tải mô hình và tokenizer đã huấn luyện
model_name = "./trained_model"
tokenizer_name = "./trained_tokenizer"
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

def predict(ticker, today_price, sentiment_label, sentiment_score, news_title):
    # Tạo input text bao gồm cảm xúc và tin tức
    input_text = (
        f"Dự báo: Giá hôm nay của {ticker} là {today_price}. "
        f"Tin tức: {news_title}. Cảm xúc: {sentiment_label} (Điểm: {float(sentiment_score):.2f})"
    )

    # Tokenize input
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True, max_length=256)

    # Dự báo với mô hình đã huấn luyện
    model.eval()  # Chuyển mô hình sang chế độ đánh giá
    with torch.no_grad():
        output = model.generate(inputs['input_ids'], max_length=20, num_beams=5, early_stopping=True)

    # Giải mã kết quả dự báo
    predicted_output = tokenizer.decode(output[0], skip_special_tokens=True)

    # Tách kết quả thành 3 phần (delta1, delta2, delta3)
    delta1, delta2, delta3 = map(float, predicted_output.split(','))

    # Tính giá cổ phiếu trong 3 ngày tới
    predicted_price_day1 = today_price + delta1
    predicted_price_day2 = today_price + delta2
    predicted_price_day3 = today_price + delta3

    return predicted_price_day1, predicted_price_day2, predicted_price_day3
```

```python
# Ví dụ sử dụng hàm dự báo
ticker = "FPT"
today_price = 136000  # Giá hôm nay
sentiment_label = "Tiêu cực"  # Cảm xúc: tiêu cực
sentiment_score = 0.75  # Điểm cảm xúc
news_title = "Cổ phiếu FPT diễn biến khó lường"

# Dự báo 3 ngày tới
predicted_prices = predict(ticker, today_price, sentiment_label, sentiment_score, news_title)
print(f"Dự báo giá trong 3 ngày tới: {predicted_prices}")
```
